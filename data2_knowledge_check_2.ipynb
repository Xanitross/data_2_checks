```python
import pandas as pd
```

Find and access data

```python
df = pd.read_csv("https://github.com/Xanitross/data_2_checks/blob/main/bikes_data.csv?raw=true")
df
```

Fix Two Issues with the data. 

The first thing we will do is convert the column names into snake_case

```python
df.columns = [c.lower().replace(' ', '_') for c in df.columns]
df
```

The "Country of Origin", "Number of Seating", and "Number of Cylinders" columns are fairly wordy, let's shorten that.

```python
df.rename(columns = {"country_of_origin":"country", "number_of_seating":"seats", "number_of_cylinders":"cylinders"}, inplace=True)
df
```

Another issue that I see is that the pricing is in INR (Indian Rupees) and that some items give the full cost while others list the price in lakhs (one hundred thousands), let's start by cutting the column name down to just "price"

```python
df["price"] = df["price_(in_inr)"]
df.drop("price_(in_inr)", inplace=True, axis=1) 
df
```

```python
df.columns
```

Now let's put that column back into the middle of the df, where it was before

```python
df = df[['company', 'country', 'model', 'number_of_cc', 'horsepower', 'torque',
       'transmission_type', 'drivetrain', 'seats', 'price', 'year', 'looks',
       'body_type', 'engine_type', 'cylinders']]
df
```

Now to identify which columns use lakhs

```python
df[df.price.str.contains('lakh')]
```

Now we are going to split those rows into a seperate DF to work on them

```python
df2 = df[df.price.str.contains('lakh')]
df2
```

And remove them from the original DF

```python
df = df[df.price.str.contains('lakh') == False] 
df
```

Now we need to remove all of the characters and convert from a string to a float

```python
df2["price"] = df2["price"].str.replace(r'[^.0-9]', '')
df2["price"] = df2["price"].astype(float)
df2
```

Now we will multiple each of the lahk prices by 100,000 to match the other prices

```python
df2.price *= 100000
df2
```

In order to re-add the lakh rows to the main DF we'll need to first update the price column in the main DF to also be a float

```python
df["price"] = df["price"].str.replace(r'[^.0-9]', '')
df["price"] = df["price"].astype(float)
df
```

```python
df = df.merge(df2, how="outer")
df
```

Now with the dataframe back together we can see what the prices would be in USD instead, as of 7/13/2023 1 USD is worth 82 INR so we will divide the current prices by 82 (rounded to 2 decimal places) to get our answer

```python
INR=82
df["price"]=(df["price"]/INR).round(2)
df
```

Let's now format that column for USD to make it more readable

```python
df['price'] = df["price"].map("${:,.2f}".format)
df
```
